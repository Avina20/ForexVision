##Downloading the data and storing it in both databases

```python
currency_pairs = ['EUR/USD', 'GBP/EUR', 'USD/CHF']
```

```python
max_N_tracker = {}
max_N_tracker["EUR/USD"] = 0
max_N_tracker["GBP/EUR"] = 0
max_N_tracker["USD/CHF"] = 0

max_vol = {}
max_vol["EUR/USD"]=0
max_vol["GBP/EUR"]=0
max_vol["USD/CHF"]=0
```

```python
# Queue for thread-safe data handling
data_queue = Queue()
```

```python
stats_summary = {
    "EUR/USD": {
        "max": 999,
        "min": -1,
        "mean": -1,
        "sum": 0,
        "count": 0,
        "cross_count": 0,
        "upper": None,
        "lower": None,
        "prev": 0,
        "keltner_bands": [],
        },
    "GBP/EUR": {
        "max": 999,
        "min": -1,
        "mean": -1,
        "sum": 0,
        "count": 0,
        "cross_count": 0,
        "upper": None,
        "lower": None,
        "prev": 0,
        "keltner_bands": [],
        },
    "USD/CHF": {
        "max": 999,
        "min": -1,
        "mean": -1,
        "sum": 0,
        "count": 0,
        "cross_count": 0,
        "upper": None,
        "lower": None,
        "prev": 0,
        "keltner_bands": [],
        },
}
```

```python
stats = stats_summary["EUR/USD"]
pair = "EUR/USD"
max_val = stats["max"]
min_val = stats["min"]
mean_val = stats["mean"]

vol = (max_val - min_val) / mean_val
max_vol[pair] = max(max_vol[pair], vol)
norm_vol = vol/max_vol[pair]

max_N_tracker[pair] = max(max_N_tracker[pair], stats["cross_count"])

N = stats["cross_count"]
if(N!=0):
  fd_norm = N/max_N_tracker[pair]

keltner_bands = []
for n in range(1, 1001):
    offset = n * 0.00001 * stats["mean"]
    upper = stats["mean"] + offset
    lower = stats["mean"] - offset
    keltner_bands.append((upper, lower))

# Save bands into stats_summary for use in next 6-min period
stats["keltner_bands"] = keltner_bands
```

```python
def fetch_fx_data(api_key, currency_pair):
    from_currency, to_currency = currency_pair.split("/")
    url = f"https://api.polygon.io/v1/conversion/{from_currency}/{to_currency}"
    params = {'apiKey': api_key}
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        try:
            data_queue.put(data)
        except Exception as e:
            print(f"Data queue error for {currency_pair}:", e)
    else:
        print(f"Error fetching data for {currency_pair}:", response.status_code, response.text)
```

```python
 DB writer thread to process data from queue
def continuous_store_data():
    while True:
        session = Session()
        data = data_queue.get()
        try:
            print("data", data)
            currency_pair = data['symbol']
            value = data['last']['ask']
            timestamp =  data['last']['timestamp']

            print("Calculate real-time stats")

            if stats_summary[currency_pair]["max"] == 999 or value > stats_summary[currency_pair]["max"]:
                stats_summary[currency_pair]["max"] = value

            if stats_summary[currency_pair]["min"] == -1 or value < stats_summary[currency_pair]["min"]:
                stats_summary[currency_pair]["min"] = value

            stats_summary[currency_pair]["sum"] += value
            stats_summary[currency_pair]["count"] += 1
            mean_val = stats_summary[currency_pair]["sum"] / stats_summary[currency_pair]["count"]

            stats_summary[currency_pair]["mean"] = mean_val

            bands = stats_summary[currency_pair].get("keltner_bands", [])

            if bands is not None:
              prev = stats_summary[currency_pair]["prev"]
              low = min(prev, value)
              high = max(prev, value)

              # Count number of bands between prev and current
              band_cross_count = 0
              for upper, lower in stats_summary[currency_pair].get("keltner_bands", []):
                  if low < upper < high or low < lower < high:
                      band_cross_count += 1

            stats_summary[currency_pair]["cross_count"] += band_cross_count

            stats_summary[currency_pair]["prev"] = value

            print("Insert into SQLite aux")
            session.execute(
                table_sql_aux.insert().values(
                    currencies=currency_pair,
                    rate=value,
                    created_on= int(datetime.utcnow().timestamp()),
                    timestamp=timestamp
                )
            )
            session.commit()

            print("Insert into MongoDB aux")
            mongo_aux.insert_one({
                'currencies': currency_pair,
                'rate': value,
                'created_on': int(datetime.utcnow().timestamp()),
                'timestamp': timestamp
            })

            print(f"Stored aux data for {currency_pair}")

        except Exception as e:
            print("DB Write Error:", e)
            session.rollback()
        finally:
            session.close()
            data_queue.task_done()
```

```python
# Aggregator every 6 minutes
def aggregate_6_min_data():
    session = Session()
    now = int(time.time())

    for pair, stats in stats_summary.items():
            try:
                print("Aggrehate for pair ", pair)
                max_val = stats["max"]
                min_val = stats["min"]
                mean_val = stats["mean"]

                vol = (max_val - min_val) / mean_val
                max_vol[pair] = max(max_vol[pair], vol)
                norm_vol = vol/max_vol[pair]

                max_N_tracker[pair] = max(max_N_tracker[pair], stats["cross_count"])

                N = stats["cross_count"]
                if(N!=0):
                  fd_norm = N/max_N_tracker[pair]

                keltner_bands = []
                for n in range(1, 1001):
                    offset = n * 0.00001 * stats["mean"]
                    upper = stats["mean"] + offset
                    lower = stats["mean"] - offset
                    keltner_bands.append((upper, lower))

                # Save bands into stats_summary for use in next 6-min period
                stats["keltner_bands"] = keltner_bands



                print("Insert into SQLite final")
                session.execute(table_sql_fin.insert().values(
                    currencies=pair,
                    max=max_val,
                    min=min_val,
                    mean=mean_val,
                    norm_vol=vol,
                    fd=fd_norm,
                    created_on=now,
                    timestamp=now
                ))
                session.commit()

                print("Insert into MongoDB final")
                mongo_final.insert_one({
                    'currencies': pair,
                    'max': max_val,
                    'min': min_val,
                    'mean': mean_val,
                    'normalized_vol': vol,
                    'fd': fd_norm,
                    'created_on': now,
                    'timestamp': now
                })

                stats["max"] = 999
                stats["min"] = -1
                stats["mean"] = -1
                stats["sum"] = 0
                stats["count"] = 0
                stats["cross_count"] = 0

            except Exception as e:
                print("Aggregation Error for", pair, ":", e)
                session.rollback()
    session.close()
```

```python
# Thread to write to DB
db_thread = Thread(target=continuous_store_data, daemon=True)
db_thread.start()

# RepeatedTimer class
class RepeatedTimer(object):
    def __init__(self, interval, function, *args, **kwargs):
        self._timer = None
        self.interval = interval
        self.function = function
        self.args = args
        self.kwargs = kwargs
        self.is_running = False
        self.start()

    def _run(self):
        self.is_running = False
        self.start()
        self.function(*self.args, **self.kwargs)

    def start(self):
        if not self.is_running:
            self._timer = Timer(self.interval, self._run)
            self._timer.start()
            self.is_running = True

    def stop(self):
        self._timer.cancel()
        self.is_running = False

api_key = 'beBybSi8daPgsTp5yx5cHtHpYcrjp5Jq'
currency_pairs = ['EUR/USD', 'GBP/EUR', 'USD/CHF']

# Start RepeatedTimers for fetching data every 6 minutes
timers = []
for pair in currency_pairs:
    timer = RepeatedTimer(1, fetch_fx_data, api_key, pair)
    timers.append(timer)

# Aggregator timer every 6 minutes
taggr = RepeatedTimer(60, aggregate_6_min_data)

# Run for 5 hours
try:
    time.sleep(60 * 5)
finally:
    for timer in timers:
        timer.stop()
    taggr.stop()

# Wait until queue is empty before finishing
while not data_queue.empty():
    time.sleep(1)
print("Data collection complete.")
```
