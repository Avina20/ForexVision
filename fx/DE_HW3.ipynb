```python
def date_range(last_start=None, earliest_end=None):
    """
    Analyze the datasets in the specified directory.
    """
    path = os.path.join(os.getcwd(), 'Data')
    files = []
    for root, _, filenames in os.walk(path):
        for f in filenames:
            if f.endswith('.csv'):
                files.append(os.path.relpath(os.path.join(root, f), path))
    
    
    for filename in files:
        file_path = os.path.join(path, filename)
        print(os.path.splitext(os.path.basename(file_path))[0])
        df = pd.read_csv(file_path)
        df['Date'] = pd.to_datetime(df['datetime'])
        start_date = df['Date'].min()  
        end_date = df['Date'].max()
        earliest_end = end_date if earliest_end is None else min(earliest_end, end_date)
        last_start = start_date if last_start is None else max(last_start, start_date)
        print(f"Date range: {start_date} to {end_date}")
    
    print(f"Overall date range: {last_start} to {earliest_end}")
```

```python
COMMON_START = '2021-01-08 00:00:00'
COMMON_END = '2023-01-24 23:00:00'
```

```python
def process_single_file(file_args):
    """
    Process a single CSV file (thread worker function)
    Returns: (currency_pair, records) or None if error
    """
    
    file_path, currency_pair =file_args
    try:
        # Read CSV with specific columns and error handling
        df = pd.read_csv(file_path, usecols=['vw', 'n', 'datetime'])
        
        # Convert and filter datetime
        df['datetime'] = pd.to_datetime(df['datetime'])
        df = df[(df['datetime'] >= COMMON_START) & 
                (df['datetime'] <= COMMON_END)]
        
        if df.empty:
            return []
        
        # Resample to hourly data and sort it as the dates are not in order
        df.set_index('datetime', inplace=True)
        df = df.sort_index()
        
        hourly_df = df.resample('1H').agg({
            'vw': 'last',      # Use last VWAP of the hour
            'n': 'sum'         # Sum of transactions for liquidity measure
        }).rename(columns={'vw': 'vwap', 'n': 'liquidity'})
        
        # Remove hours with no transactions
        hourly_df = hourly_df[hourly_df['liquidity'] > 0]
        
        # Prepare data for database
        records = []
        for timestamp, row in hourly_df.iterrows():
            records.append((
                timestamp.strftime('%Y-%m-%d %H:%M:%S'),  # timestamp
                currency_pair,                            # currency_pair
                row['vwap'],                              # vwap
                int(row['liquidity'])                    # liquidity
            ))
        
        return records
    
    except Exception as e:
        print(f"Error processing {file_path}: {str(e)}")
        return []
```

```python
def process_all_files():
    """Main function to process all CSV files using multithreading"""
    data_path = os.path.join(os.getcwd(), 'Data')
    csv_files = []
    
    # Find all CSV files and their currency pairs
    for root, _, files in os.walk(data_path):
        for file in files:
            if file.endswith('.csv'):
                currency_pair = os.path.splitext(file)[0].upper()
                file_path = os.path.join(root, file)
                csv_files.append((file_path, currency_pair))
    
    
    # Process files in parallel
    with ThreadPoolExecutor(max_workers=8) as executor:
        records = list(tqdm(executor.map(process_single_file, csv_files), total=len(csv_files), desc="Processing files"))
    
    # Batch insert records to databases
    total_records = 0
    for record in tqdm(records, desc="Storing in databases"):
        if record is None:
            continue
            
        total_records += len(record)
        
        # SQLite insert with timestamp
        sql_records = [r[:4] for r in record]
        if not aux_db.bulk_insert_sqlite(sql_records):
            print(f"Failed to insert {len(sql_records)} records")
        
        
        # Batch insert to MongoDB
        mongo_docs = [{
            "timestamp": r[0],
            "currency_pair": r[1],
            "vwap": r[2],
            "liquidity": r[3],
            "insert_timestamp": datetime.now()
        } for r in record]
        
        # Insert in batches of 1000
        for i in range(0, len(mongo_docs), 1000):
            batch = mongo_docs[i:i + 1000]
            if not aux_db.bulk_insert_mongo(batch):
                print(f"Failed to insert batch {i//1000} for {currency_pair} to MongoDB")
    
    print(f"\nProcessing complete. Total records inserted: {total_records}")
```

```python
def gen_keltner_bands(mean):
    """Generate Keltner Bands for a given mean value"""
    keltner_bands = []
    for n in range(1, 1001):
        offset = n * 0.00001 * mean
        upper = mean + offset
        lower = mean - offset
        keltner_bands.append(upper)
        keltner_bands.append(lower)
    return keltner_bands
```

```python
import bisect

def count_crossings(bands, values):
    """
    Count Keltner band crossings using position tracking.
    
    Parameters:
    - bands: List of Keltner band values (upper and lower)
    - values: List of price values
    
    Returns:
    - Total number of bands crossed during price movements
    """
    if len(values) < 2 or not bands:
        return 0
    
    # Get unique sorted band values
    sorted_bands = sorted(set(bands))
    crossed_count = 0
    
    # Get initial position
    prev_pos = bisect.bisect_right(sorted_bands, values[0])
    
    for i in range(1, len(values)):
        curr_pos = bisect.bisect_right(sorted_bands, values[i])
        
        # Count crossed bands (absolute difference between positions)
        crossed_count += abs(curr_pos - prev_pos)
        prev_pos = curr_pos
    
    return crossed_count
```

```python
def calculate_6h_features(currency_pair, aux_db, main_db):
    """
    Calculate all 6-hour features with proper band tracking and error handling
    
    Parameters:
    - currency_pair: String representing the currency pair (e.g., 'EURUSD')
    - aux_db: Database connection to auxiliary data
    - main_db: Database connection to main data
    """
    # Get hourly data with error handling
    try:
        hourly_data = aux_db.find_records(
            query={"currency_pair": currency_pair},
            sort=[("timestamp", 1)]  # Ascending order
        )
    except Exception as e:
        print(f"Error fetching data for {currency_pair}: {str(e)}")
        return

    if not hourly_data:
        print(f"No hourly data found for {currency_pair}")
        return
    
    # Convert to DataFrame with validation
    try:
        df = pd.DataFrame(hourly_data)
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df.set_index('timestamp', inplace=True)
        
        # Validate required columns exist
        if not all(col in df.columns for col in ['vwap', 'liquidity']):
            print(f"Missing required columns in data for {currency_pair}")
            return
    except Exception as e:
        print(f"Error processing DataFrame for {currency_pair}: {str(e)}")
        return

    # Initialize tracking variables
    global_max_vol = 0
    global_max_cross = 0
    prev_mean = None
    keltner_bands = None
    all_rows = []
    
    # Process in 6-hour chunks
    for period_end, group in df.resample('6H'):
        try:
            # Skip empty groups
            if group.empty:
                print(f"Empty 6-hour period for {currency_pair} at {period_end}")
                continue
                
            # Calculate basic features
            row = {
                'timestamp': period_end.strftime('%Y-%m-%d %H:%M:%S'),
                'currency_pair': currency_pair,
                'vwap': group['vwap'].iloc[-1],  # Last price
                'max': group['vwap'].max(),
                'min': group['vwap'].min(),
                'mean': group['vwap'].mean(),
                'avg_liquidity': group['liquidity'].mean(),
            }
            
            # Calculate volatility
            current_vol = (row['max'] - row['min']) / row['mean']
            global_max_vol = max(global_max_vol, current_vol)
            row['norm_vol'] = current_vol / global_max_vol if global_max_vol > 0 else 0
            
            # Calculate fractal dimension (after first period)
            if prev_mean is not None:
                crossings = count_crossings(keltner_bands, group['vwap'].values)
                global_max_cross = max(global_max_cross, crossings)
                row['fd'] = crossings / global_max_cross if global_max_cross > 0 else 0
            else:
                row['fd'] = 0.0
            
            # Update bands for next period
            current_mean = row['mean']
            keltner_bands = gen_keltner_bands(current_mean)
            prev_mean = current_mean
            
            all_rows.append(row)
            
        except Exception as e:
            print(f"Error processing {currency_pair} at {period_end}: {str(e)}")
            continue

    # Prepare database records
    records = []
    for result in all_rows:
        try:
            records.append((
                result['timestamp'],
                result['currency_pair'],
                float(result['vwap']),
                float(result['avg_liquidity']),
                float(result['norm_vol']),
                float(result['max']),
                float(result['min']),
                float(result['fd']),
                datetime.now().isoformat()
            ))
        except Exception as e:
            print(f"Error formatting record for {result['timestamp']}: {str(e)}")
            continue
    
    # Store in databases
    if records:
        try:
            # SQLite insertion
            if not main_db.bulk_insert_sqlite(records):
                print(f"Failed to insert {len(records)} records to SQLite for {currency_pair}")
            
            # MongoDB insertion
            mongo_docs = [{
                "timestamp": r[0],
                "currency_pair": r[1],
                "vwap": r[2],
                "avg_liquidity": r[3],
                "norm_vol": r[4],
                "max": r[5],
                "min": r[6],
                "fd": r[7],
                "insert_timestamp": r[8],
            } for r in records]
            
            if not main_db.bulk_insert_mongo(mongo_docs):
                print(f"Failed to insert records to MongoDB for {currency_pair}")
                
        except Exception as e:
            print(f"Database insertion failed for {currency_pair}: {str(e)}")
```

```python
def process_all_currency_pairs():
    """Process all currency pairs for feature extraction"""
    
    
    # Get all unique currency pairs
    currency_pairs = aux_db.collection.distinct("currency_pair")
    
    # Process each currency pair
    for cp in tqdm(currency_pairs, desc="Processing currency pairs"):
        calculate_6h_features(cp, aux_db, main_db)
    
    print("Feature extraction complete!")
```

```python
import sqlite3
import pandas as pd

def export_sqlite_to_csv(db_path, output_folder):
    """Export all tables from SQLite to separate CSV files"""
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    
    # Get all table names
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()
    
    for table in tables:
        table_name = table[0]
        df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
        df.to_csv(f"{output_folder}/{table_name}.csv", index=False)
        print(f"Exported {table_name} ({len(df)} rows) to {output_folder}/{table_name}.csv")
    
    conn.close()

# Usage
export_sqlite_to_csv('fx_main.db', 'Data')
```
